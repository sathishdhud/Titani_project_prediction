```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
df = pd.read_csv('/kaggle/input/titanic/Titanic-Dataset.csv')
knn = KNeighborsClassifier(n_neighbors=1)
d=df.loc[:,:].dropna()

d.Survived = d.Survived.map({0:'S',1:'N'})
d.Sex = d.Sex.map({'female':0,'male':1})
d.Embarked = d.Embarked.map({'S':1,'C':0})

x=d[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare']].values
y = d.Survived.values
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
x=x
knn.fit(x_train,y_train)

pred = knn.predict(x_test)
pred
```
